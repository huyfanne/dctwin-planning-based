```python
import matplotlib.pyplot as plt
import json
import pandas as pd
```

```python
dctwin_data = pd.read_csv("log/dctwin_output/episode-1/dctwin_output.csv")
eplus_data = pd.read_csv("log/eplus_output/episode-1/eplusout.csv")
with open("configs/dt/device_key_map.json", "r") as f:
    device_key_map = json.load(f)
```

```python
total_chilled_water_flow_rate_eplus = eplus_data[
    "DATA HALL 1A ACU-1 WATER INLET NODE:System Node Mass Flow Rate [kg/s](TimeStep)"
]
total_chilled_water_flow_rate_dctwin = dctwin_data[
    "chilled water loop demand inlet branch:water_mass_flow_rate"
]

plt.figure(dpi=100)
plt.plot(total_chilled_water_flow_rate_dctwin, label="DCTwin")
plt.plot(total_chilled_water_flow_rate_eplus, label="EnergyPlus")
plt.xlabel("Time Step")
plt.ylabel("Mass Flow Rate [kg/s]")
plt.title("Data Hall ACU-1 Chilled Water Flow Rate")
plt.grid(axis="y")
plt.ylim(0, 2)
plt.legend()
```

```python
chiller_power_eplus = eplus_data["CHILLER 1:Chiller Electricity Rate [W](TimeStep)"]
chiller_power_dctwin = dctwin_data["chiller 1:power"]
plt.figure(dpi=100)
plt.plot(chiller_power_dctwin / 1000, label="DCTwin")
plt.plot(chiller_power_eplus / 1000, label="EnergyPlus")
plt.xlabel("Time Step")
plt.ylabel("Power [kW]")
plt.title("Chiller-1 Power")
plt.grid(axis="y")
plt.legend()
```

```python
acu_power_dctwin = dctwin_data["Data Hall 1A acu-1:fan_power"]
acu_power_eplus = eplus_data[
    "DATA HALL 1A ACU-1 FAN:Fan Electricity Rate [W](TimeStep)"
]
plt.figure(dpi=100)
plt.plot(acu_power_eplus / 1000, label="DCTwin")
plt.plot(acu_power_eplus / 1000, label="EnergyPlus")
plt.xlabel("Time Step")
plt.ylabel("Power [kW]")
plt.title("Data Hall ACU-1 Power")
plt.grid(axis="y")
plt.legend()
```

```python
chwp_power_dctwin = dctwin_data["chiller 1 chwp:power"]
chwp_power_eplus = eplus_data["CHILLER 1 CHWP:Pump Electricity Rate [W](TimeStep)"]
plt.figure(dpi=100)
plt.plot(chwp_power_dctwin / 1000, label="DCTwin")
plt.plot(chwp_power_eplus / 1000, label="EnergyPlus")
plt.xlabel("Time Step")
plt.ylabel("Power [kW]")
plt.title("Chilled Water Pump Power")
plt.grid(axis="y")
plt.ylim(0, 1)
plt.legend()
```

```python
chiller_clg_load_dctwin = dctwin_data["chiller 1:cooling_load"]
chiller_clg_load_eplus = eplus_data[
    "CHILLER 1:Chiller Evaporator Cooling Rate [W](TimeStep)"
]

plt.figure(dpi=100)
plt.plot(chiller_clg_load_dctwin / 1000, label="DCTwin")
plt.plot(chiller_clg_load_eplus / 1000, label="EnergyPlus")
plt.xlabel("Time Step")
plt.ylabel("Chiller-1 Cooling Load [kW]")
plt.title("Total Chiller Cooling Load")
plt.legend()
plt.ylim(0, 200)
plt.grid(axis="y")
```
