Welcome to DCWiz-Model-Library! In this tutorial, we will show you how to use the library to build a data center digital twin. The digital twin can be used to simulate the data center performance and optimize the data center operation. We will guide you through the following steps:

* Define the models needed for the digital twin. The models serve as the building blocks of the digital twin. A model will describe a specific aspect of a device in a data center (e.g., the power model of a server describe its rated power, cpu power consumption etc). The models are defined in the `models` folder.
* Define the data center building envelop geometry. For a data center, it might be host inside a commercial building or a dedicated building. The geometry info describes the size of the data center building. 
* Define the construction of a data center. A data center is composed of data halls and the supporting chiller plants. Therefore, in the construction, we need to define how each data hall is constructed, and how the chiller plant is built.

## Define the models
In this section, we will define the models needed for the digit twin. The models can be categorized into the following categories:
* __Computing Models__:
   * __Server Model__: describe the computing hardware parameters (cpu, memory, storage etc).
* __Cooling Models__:
   * __ACU Cooling Model__: describe the ACU cooling-related parameters (rated flow rate etc).
   * __Chiller Cooling Model__: describe the chiller cooling-related parameters (reference capacity etc).
   * __Cooling Tower Cooling Model__: describe the cooling tower cooling-related parameters (reference capacity etc).
   * __Pump Cooling Model__: describe the pump cooling-related parameters (maximum flow rate etc).
   * __Pipe Cooling Model__: describe the pipe cooling-related parameters (pipe diameter etc).
   * __Server Cooling Model__: describe the server fan volumetric flow rate ratio
* __Power Models__:
   * __Server Power Model__: describe the server power-related parameters (rated power, cpu power consumption etc).
   * __ACU Power Model__: describe the ACU fan power-related parameters (rated power, rated pressure drop etc).
   * __Chiller Power Model__: describe the chiller power-related parameters (rated COP etc).
   * __Cooling Tower Power Model__: describe the cooling tower power-related parameters (rated power etc).
   * __Pump Power Model__: describe the pump power-related parameters (rated power etc).
* __Construction Models__:
   * __Surface Material Model__: describe the surface material-related parameters (thermal conductivity etc).
* __Curve Models__:
   * __Quadratic Curve Model__: define a parametric quadratic curve.
   * __Cubic Curve Model__: define a parametric cubic curve.
   * __Bi-quadratic Curve Model__: define a parametric bi-quadratic curve.
* __Electrical Models__:
   * __PV Performance Model__: describe the PV performance-related parameters (number modules, rated efficiency etc).
   * __Battery Performance Model__: describe the battery performance-related parameters (rated capacity, rated efficiency etc).
   * __Transformer Model__: describe the transformer performance-related parameters (efficiency etc).
   * __Inverter Model__: describe the inverter performance-related parameters (efficiency etc).
* __Heat Gain Models__: 
   * __People Heat Gain Model__: describe the people heat gain-related parameters (heat gain per person etc).
   * __Lighting Heat Gain Model__: describe the lighting heat gain-related parameters (heat gain per light per squared meter etc).
   * __ITE Equipment Heat Gain Model__: describe the ITE equipment heat gain-related parameters (number units, Watt per unit etc).
   * __Electrical Equipment Heat Gain Model__: describe the electrical equipment heat gain-related parameters (Watts per squared meter, rated power). ITE is a special case of electrical equipment. Electrical equipment is a general term for all electrical equipment in a data center, excluding ITE. It might include electrical equipments inside the office etc.

## Define the chiller plant in a data center
In this section, we will define the chiller plant in a data center. A chiller plant is composed of chillers, cooling towers and pumps. A typical chiller plant can be modeled as several chilled water loops and condenser water loops.  Each loop can be modeled with several supply-side branches and demand-side branches. The supply-side branches are used to define the components on the supply-side of the chilled water loop, while the demand-side branches are used to define the components on the demand-side of the chilled water loop. For a loop, supply-side devices provide cooling energy to the demand-side. The demand-side devices request for cooling energy from the supply-side. 

# Define chilled water loop
